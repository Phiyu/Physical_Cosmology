To plot 
$$
    \Omega(t)-1= \dfrac{\Omega_{k,0}}{\Omega_{\rm m,0}a^{-1}+\Omega_{\rm r,0}a^{-2}+\Omega_{k,0}},
$$
we can just simply suppose $\Omega_{\rm m,0} = 0.299, \Omega_{\rm r,0 = 0.001}$ and $\Omega_{k,0} = 0.7$.

```python
import numpy

def func(a):
    Omm = 0.299
    Omr = 0.001
    Omk = 0.7
    return Omk / (Omm * a**-1 + Omr * a**-2 + Omk)

import matplotlib.pyplot as plt
plt.style.use('default')

a = numpy.logspace(-40, 1, 1000)
plt.plot(a, func(a))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Scale Factor a')
plt.ylabel(r'$\Omega(t) - 1$')
plt.title(r'Evolution of $\Omega(t) - 1$ with Scale Factor a')
plt.grid(True)
plt.show()
```
