```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
plt.style.use('default')

def T(x):
    return (np.log10(1+2.34*x))/(np.log10(np.e) * (2.34*x)) * ((1 + 3.89*x + (16.1*x)**2 + (5.46*x)**3 + (6.71*x)**4)**(-0.25))

def window(x):
    j1 = np.sin(x) / x**2 - np.cos(x) / x
    return 3 * j1 / x

def integral(x):
    w = window(x * 8)
    A = 2.1e-9
    factor = 4 / 25 / (70**4) * A
    Delta = factor * (x**4) * (T(x/0.7**2)**2)
    return Delta * Delta * w * w / x

k = np.logspace(-4, 100, 10000)
integrand = integral(k)

plt.plot(np.log(k)/np.log(np.e), integrand)
plt.xlabel(r'$\ln k \, [h \, \mathrm{Mpc}^{-1}]$')
plt.ylabel(r'Integrand for $\sigma_8^2$')
plt.xscale('log')
plt.yscale('log')
plt.title(r'Integrand for $\sigma_8^2$ vs. $\ln k$')
plt.grid()
plt.show()
```
